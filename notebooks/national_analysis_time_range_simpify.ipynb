# 2. define some functions

```python
def get_data (startyear, endyear):
    yearlist=range(startyear,endyear+1)
    subset_data=subset_urban[subset_urban['year'].isin(yearlist)] 
    subset_counts = subset_data.groupby('NAME10')['ID_V4.02'].count().to_frame(name='Total # Samples').reset_index()
    urban_rank=subset_counts.sort_values(by='Total # Samples',ascending=False)
    cities=urban_rank.iloc[:50]['NAME10'].tolist()  ## 50 is the number of cities of interest
    subset_cities= subset_urban[subset_urban['NAME10'].isin(cities)]
    pollutant_counts=subset_cities.groupby('NAME10').count()
    nonzero_cols= pollutant_counts.dropna(axis=1, how='any')
    pollutantcount=nonzero_cols[pollutants].reset_index()  ## count with only pollutants
    pollutantcount_nonzero=pollutantcount.astype(bool).sum(axis=0)
    sorted_pollutants=pollutantcount_nonzero.to_frame().sort_values(by=0, ascending=False)
    pollutant_export=pollutantcount[list(sorted_pollutants.T)[:15]] ## 15 pollutants should be enough for the maximum
    return pollutant_export
```

```python
def get_sort(startyear, endyear):
    pollutant_export=get_data(startyear, endyear)
    pollutantlist=list(pollutant_export)[1:]
    national_pollutants_n=pollutant_export[pollutantlist].astype(bool).sum(axis=0)
    pollutants_sort=national_pollutants_n.to_frame().index
    return pollutants_sort
```

```python
def getNcity(Npollutants,startyear, endyear):
    pollutant_export=get_data(startyear, endyear)
    pollutants_sort=get_sort(startyear, endyear)
    pollutants_of_interest=list(pollutants_sort[:Npollutants])
    data_temp=pollutant_export[['NAME10']+ pollutants_of_interest]
    data_temp_nonzero=data_temp.replace(0,pd.np.nan).dropna(axis=0,how='any').reset_index().drop(['index'],axis=1)
    Ncity=len(data_temp_nonzero)
    return Ncity
```

```python
## this is the final function
def get_output(startyear, endyear):
    pollutants_sort=get_sort(startyear, endyear)
    Npollutants_list=range(1,len(pollutants_sort)+1)
    Ncity_list=[getNcity(i,startyear, endyear) for i in Npollutants_list]
    pollutants_dict={'Ncity':Ncity_list,'pollutants_added':pollutants_sort,'Npollutants':Npollutants_list}
    pollutants_ncity=pd.DataFrame(pollutants_dict)
    return pollutants_ncity
```

# 4. further judge is based on number of samples and stations, conclusion is that 1992-2003 is better

```python
pollutants=['Phosphorous Total as P (mg/L)','Zinc Total (ug/L)','Lead Total After 1984 (ug/L)','TSS (mg/L)',
'Copper Total (ug/L)','Nitrogen Kjeldahl Total (mg/L)','BOD5 as O2 (mg/L)']
attributes=['State','Station_Name','NAME10','Drainage_Area_Acres','Principal Landuses' ,'date','year','Lat','Lon']
```

```python
def get_export_data(startyear,endyear):    
    subset_urban_se=subset_urban[pollutants+attributes]
    yearlist=range(startyear,endyear+1)
    subset_urban_temp=subset_urban_se[subset_urban['year'].isin(yearlist)] 
    subset_urban_se2= subset_urban_temp.dropna(axis=0, how='any')
    export=subset_urban_se2[['Station_Name','year','Lat','Lon','NAME10','Drainage_Area_Acres']].drop_duplicates().reset_index(drop=True)   
    return export
```

```python
def get_export_data2(startyear,endyear):    
    subset_urban_se=subset_urban[pollutants+attributes]
    yearlist=range(startyear,endyear+1)
    subset_urban_temp=subset_urban_se[subset_urban['year'].isin(yearlist)] 
    subset_urban_se2= subset_urban_temp.dropna(axis=0, how='any')
    export=subset_urban_se2[['Station_Name','Lat','Lon','NAME10','Drainage_Area_Acres',]].drop_duplicates().reset_index(drop=True)   
    return export
```

```python
def get_descriptive(startyear,endyear):    
    subset_urban_se=subset_urban[pollutants+attributes]
    yearlist=range(startyear,endyear+1)
    subset_urban_temp=subset_urban_se[subset_urban['year'].isin(yearlist)] 
    subset_urban_se2= subset_urban_temp.dropna(axis=0, how='any')
    sample_count_bystation=subset_urban_se2[['NAME10','Station_Name','State']].groupby(['NAME10','Station_Name']).count().reset_index()
    sample_count_bycity=sample_count_bystation[['NAME10','State']].groupby(['NAME10']).sum().reset_index()
    station_count=sample_count_bystation[['NAME10','Station_Name',]].groupby('NAME10').count().reset_index()
    sample_count_byyear=subset_urban_se2[['NAME10','State','year']].groupby(['NAME10','year']).count().reset_index()
    year_count_bycity=sample_count_byyear[['NAME10','year',]].groupby('NAME10').count().reset_index()
    temp=station_count.merge(sample_count_bycity,on='NAME10')
    allcount_bycity=temp.merge(year_count_bycity,on='NAME10')
    allcount_bycity.columns=['NAME10','Nstation','Nsample','Nyear']
    allcount_bycity
    
    return allcount_bycity
```
