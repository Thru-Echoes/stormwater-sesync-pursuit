# LA city

```python
## subset data to only pollutants, stations year and data columns 

ladata=subset_urban.loc[subset_urban['NAME10']=='Los Angeles--Long Beach--Anaheim, CA']
Collist=['Station_Name','year','date','NAME10']+pollutants
ladata2=ladata[Collist]
ladata2.head()
```

```python
## count null values in each pollutants columns
lanull=ladata2[pollutants].isnull().sum().to_frame().sort_values(by=0, ascending=True)
lanull.head()
```

```python
## drop columns which contain NA values larger than 300, 300 is selected after some trial and errors to ensure the 
## inclusion of some important pollutant categories, can be modified later

pollutant_la=list(lanull[lanull[0]<300].index)
print(pollutant_la)
len(pollutant_la)
```

```python
## subset la data by keeping the pollutants with fewer than 300 missing values

ladata3=ladata2[pollutant_la+['Station_Name','year','date','NAME10']]
ladata3.head()
```

```python
## drop rows with more than 5 missing values in columns, check the sample size

ladata4=ladata3[ladata3.isnull().sum(axis=1)<5]  ## tolerate at most 5 missing values
ladata4.shape
```

```python
ladata5=ladata4.reset_index().drop(['index'],axis=1)
ladata5.to_csv('lada_pollutant.csv')
```

```python
#counts number of years in the urban area

la_year_count = ladata4.groupby('year').count()['NAME10']
print(la_year_count)
```

```python
#counts number of stations in the urban area

la_station_count=ladata4.groupby('Station_Name').count()['NAME10']
print(la_station_count)
print(la_station_count.shape[0])
```
